## KMeans and Silhouette Testing

```python
from cluster import kmeans, silhouette
from cluster import (
  make_clusters, 
  plot_clusters, 
  plot_multipanel)
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
"""
Cluster Generation
"""

# here we are making tightly clustered data by lowering the scale param
t_clusters, t_labels = make_clusters(scale=0.3) 

# here we are making loosely clustered data by increasing the scale param
l_clusters, l_labels = make_clusters(scale=2) 

# here we are making many clusters by adjusting the `k` param
m_clusters, m_labels = make_clusters(k=4)

# here we are directly controlling the dimensionality of our data 
#   1000 observations 
#   200 features 
#   3 clusters)
d_clusters, d_labels = make_clusters(n=1000, m=200, k=4)

"""
Cluster Visualization
"""
# show the visualization for some clusters and their associated labels
plot_clusters(m_clusters, m_labels)

# show a multipanel visualization of clusters with true labels, predicted labels, and silhouette scores
# you will need to calculate these predicted labels with your kmeans implementation and the scores 
# with your silhouette score implementation
# instantiation of model with some params
model = kmeans.KMeans(4)
np.random.seed(1)
model.fit(m_clusters)
s_score = silhouette.Silhouette()

# predict the labels on some input data
pred_labels = model.predict(m_clusters)
scores = s_score.score(m_clusters, pred_labels)

# print the sklearn and my silhouette scores to compare
print(silhouette_score(m_clusters, pred_labels))
print(scores.mean())

# plot the multipanel visualization
plot_multipanel(m_clusters, m_labels, pred_labels, scores)
```

## Misc Scratch Testing

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs
from cluster import kmeans, silhouette
import numpy as np
from scipy.spatial.distance import cdist, pdist
```

```python
X, _ = make_blobs(random_state=42)
model = kmeans.KMeans(2)
s_score = silhouette.Silhouette()
model.fit(X)
print(silhouette_score(X, model.predict(X)))
print(s_score.score(X, model.predict(X)).mean())
```
